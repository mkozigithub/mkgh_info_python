```python
import pandas as pd
```

```python
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_sample.csv')
df.head(2)
```

### Aggregating a series

```python
# nan are automatically excluded
print(df['year'].min())
print(df['year'].max())
print(df['year'].sum())
print(df['year'].mean())
print(df['year'].median())
print(df['year'].std())
```

### Aggregating a data frame

Can use individual aggregation functions

```python
df.sum() # default to aggregating on a per column basis (axis=0)
```

Can use 'agg' function with single or multiple aggregation functions

```python
df.agg(['min','max','mean','std'])
```

Can apply 'agg' function across columns

```python
df.agg('mean', axis=1)  # apply function across columns (so, on a per row basis)
```
