```python
import pandas as pd
```

```python
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_data.csv',low_memory=False)
```

### Using .str.strip()

```python
df.loc[df['title'].str.contains('\s$', regex=True),:]  # find all rows where title ends with whitespace
```

```python
df.loc[df['title'].str.contains('\s$', regex=True),:]['title'].map(lambda x: f"'{x}'"[-25:])
```

```python
title_stripped = df['title'].str.strip() # strip whitespace from front and back of string (not an inplace operation)
title_stripped_left = df['title'].str.lstrip()  # strip whitespace from front of string
title_stripped_right = df['title'].str.rstrip() # strip whitespace from end of string
df['title'] = title_stripped
```

```python
df.loc[df['title'].str.contains('\s$', regex=True),:]  # no more titles that end in whitespace
```

### Using .transform()

```python
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_data.csv',low_memory=False)
```

```python
df.loc[df['title'].str.contains('\s$', regex=True),:]['title'].map(lambda x: f"'{x}'"[-25:])
```

```python
df['title'].transform(lambda x: x.strip())  # this does not change the source dataframe
df['title'] = df['title'].transform(lambda x: x.strip())  # actually change the source dataframe
```

```python
df.loc[df['title'].str.contains('\s$', regex=True),:]['title'].map(lambda x: f"'{x}'"[-25:])
```
