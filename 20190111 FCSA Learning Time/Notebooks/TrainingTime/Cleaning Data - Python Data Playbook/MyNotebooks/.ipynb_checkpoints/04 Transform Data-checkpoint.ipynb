```python
import pandas as pd
```

```python
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_sample.csv')
```

### Transform height column

```python
df['height'].transform(lambda x: print(x))  # note: print(x) return 'None'
```

```python
df['height'].transform(lambda x: x / 10)
```

```python
df.groupby('artist')
```

```python
df.groupby('artist').nunique()
```

```python
df.groupby('artist').transform('nunique')  # use pandas nunique method
```

```python
df.groupby('artist')['height'].transform('nunique')
```

```python
df.groupby('artist')['height'].transform('mean') # use pandas mean method
```

```python
df.groupby('artist')['height'].mean()
```

```python
df['mean_height_by_artist'] = df.groupby('artist')['height'].transform('mean')
```
