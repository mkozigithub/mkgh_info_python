```python
import pandas as pd
```

```python
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_sample.csv')
df.set_index('id',inplace=True)  # change index to make difference between .loc and .iloc more clear: label vs. integer position
df.head(3)
```

```python
df.iloc[0,:]  # iloc uses the integer position of the row, not the row label, row returned as series
```

```python
df.iloc[2:5, :]  # iloc slice is inclusive/exclusive [x:y), so x is included, but y is not
```

```python
df.loc[0:3,:]  # no rows in resulting dataframe, since there is no 0,1,2,3 in index (now index starts at 1035)
```

```python
df.loc[1035:1037,:]   # .loc slice is inclusive/inclusive [x,y]
```

### Getting Columns Using .iloc

```python
df.iloc[0:3, 0:3]  # get columns at integer position 0,1,2  (Note: id is an index and is not considered a column)
```

```python
df.reset_index(inplace=True)
df.head(3)
```

```python
df.iloc[0:3, 0:3]  # get columns at integer position 0,1,2  (Note: id is not index and is considered a column)
```

```python
df.iloc[[1,3,5],[0,2,4,6]]  # can use lists of integer positions for specific rows and columns
```
