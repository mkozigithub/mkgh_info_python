```python
df.loc[0,:]  # equivalent to df.loc[0]
```

```python
df.loc[0:2,:]  # slice for row selector, note this slice is inclusive/inclusive [0:2] and selects index value 0,1,2

#df[0:2] # note: in this case the slice is inclusive/exclusive and returns rows for index values 0,1 only
# Note: .loc uses row label and df[] uses integer row position
```

```python
df[0:2] # note: in this case the slice is inclusive/exclusive and returns rows for index values 0,1 only
```

```python
df.loc[1:3, 'title']  # rows with row labels 1,2,3 and only column 'title', returns a series
```

```python
df.loc[1:3, ['artist','title']]  # returns a new dataframe
```

```python
df.loc[[1,5],['id','artist','title']]  # ask for specific rows and columns using lists instead of slices
```

```python
df.head(1)
```

```python
df.loc[[1,4], 'artistId':'year']  # can use a slice on columns, works in order of columns as stored in df.columns
```

### Filtering with .loc

```python
df.loc[df['artist'] == 'Blake, Robert', :]  # retrun new dataframe with all columns where row artist is Blake, Robert
```

```python
df.loc[df['artist'] == 'Blake, Robert', ['id','artist','year']] 
```

```python
df.loc[(df['artist'] == 'Blake, Robert') & (df['year'] > 1750), ['id','artist','year']]  # combining criteria with &
```
