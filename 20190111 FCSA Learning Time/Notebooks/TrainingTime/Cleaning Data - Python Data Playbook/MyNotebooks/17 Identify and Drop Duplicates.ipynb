```python
import pandas as pd
df = pd.read_csv('../cleaning-data-python-data-playbook/artwork_sample.csv', low_memory=False)
df.shape
```

```python
df.drop_duplicates().shape  # drops fully duplicated row (every column in both rows matches)
```

```python
df.drop_duplicates(subset=['artist'])  # drop duplicates where 'artist' column matched
```

```python
df.drop_duplicates(subset=['artist'],keep='first')  # keep only the first dupicated row
```

```python
df.drop_duplicates(subset=['artist'],keep='last')  # keep the last duplicated row
```

```python
df.drop_duplicates(subset=['artist'],keep=False)  # eliminate all duplicated rows, in this case all rows have at least one dup
```

```python
# to make the drop_duplicates permanent, use the inplace=True argument
df.drop_duplicates(subset=['artist'],keep='first',inplace=True)
df.shape
```
